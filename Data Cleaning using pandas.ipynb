```python
import pandas as pd
import numpy as np


'''creating raw data 

and convert to dataframe'''

# Create a sample dataset
data = {
    'Employee_ID': [101, 102, 103, 104, 105, 105, 107, 108, 109, 110,110],
    'First_Name': ['John', 'Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Helen', 'Ivy','Ivy'],
    'Last_Name': ['Smith', 'Johnson', 'Williams', 'Brown', 'Jones', 'Davis', 'Miller', 'Wilson', 'Moore', 'Lee','Lee'],
    'Department': ['HR', 'Finance', 'IT', 'Sales', 'Marketing', 'HR', 'IT', 'Finance', 'Sales', 'Marketing','Marketing'],
    'Salary': [60000, 65000, 70000, np.nan, -99999, 75000, 72000, 67000, np.nan, 68000,68000],
    'Manager_ID': [None, 302, 403, 504, 605, 706, 807, None, 1009, 1110,1110]
}

# Create a DataFrame 
df = pd.DataFrame(data)

# Display the  dataset
df
```
