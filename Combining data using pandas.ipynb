## join operation

```python
data
```

```python
data1
```

```python
#left join
# Left join
df_left=pd.merge(data, data1, on='ID', how='left')
print(df_left.shape)
df_left

```

This will include all rows from the left_df and only the matching rows from the right_df. Rows in the result that have no match in the right dataframe will have NaN values in the columns from the right dataframe.

```python
# Right join
df_right=pd.merge(data, data1, on='ID', how='right')
print(df_right.shape)
df_right
```

all the rows from the right set and only the matching rows from the left set. If there's no match in the left set for a row in the right set, you'll still get that row in the result, with NaN or null values for the columns from the left set.

```python
# Outer join
df_outer=pd.merge(data, data1, on='ID', how='outer')
print(df_outer.shape)
df_outer
```

An outer join will give you all the rows from both sets, and where there are matching values in the common column, it includes those rows as well  with NaN or null values

```python
# Inner join
df_inner=pd.merge(data, data1, on='ID', how='inner')
print(df_inner.shape)
df_inner
```

An inner join will give you only the rows where there is a match in both sets. Rows that don't have a matching value in the other set will be excluded from the result.
